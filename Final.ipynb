### To create my computer generated novel I want to create a new narrative in the style of a poem from my previously generated text.<br/>   

1 - Splitting the text.<br/> 
I split my generated text of 18289 words into 800 words. (The cell after the next one shows how I do this and save them as txt files in my folder structure.) <br/> 

2 - Generating the poem.<br/> 
I want to write a loop that will take 800 words at a time and randomly select words from these 800 words in different ways. Then join all the words together in a certain structure to generate a poem. 

We start with 1 - Splitting the text. First line of code creates a text document with the title "1.txt"

```python
f= open("1.txt","w+")
```

Second line of code, selects the words between 0 and 800 and joins them. Prints them. The writes them to the text document we saved above. 

```python
selected = words[0:800]
joined = ' '.join(selected)
print(joined)
f.write(joined)
```

The third line of code, closes the document.

```python
f.close() 
```

Now we move onto 2 - Generating the poem.<br/> The code below, selects and prints out 25 random selected words from the words 0 to 800 from my generated text.<br/> Then selects and prints out 10 randomly selected words from the words 0 to 800.<br/> Then joins them in a 6 word sentence structure.<br/> Then prints them.



```python
for i in range(25):
    for y in range(10):
        selected = random.sample(words[0:800], 6)
        joined = ' '.join(selected)
        print(joined)
```
